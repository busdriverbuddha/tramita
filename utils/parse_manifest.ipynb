```python
import json

from pathlib import Path

import pandas as pd


MANIFEST_PATH = Path("../data/bronze/snapshots/bronze-2020-2024-v2/manifest.json").expanduser()
MANIFEST_PATH.exists()
```

```python
with open(MANIFEST_PATH) as f:
    data = json.load(f)
```

```python
rows = []
for source_name, source in data['sources'].items():
    for entity_name, entity in source['entities'].items():
        for year_name, year in entity['years'].items():
            rows.append({
                'source': source_name,
                'entity': entity_name,
                'year': year_name,
                'total_rows': year['total_rows'],
                'total_bytes': year['total_bytes'],
            })
        
        
```

```python
df = pd.DataFrame(rows)
```

```python
compiled_df = df.groupby(['source', 'entity'], group_keys=True).agg('sum').reset_index().drop('year', axis=1)
compiled_df
```

```python
compiled_df.to_excel('compiled_data.xlsx', index=False)
```
